```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# CSV file named 'titanic.csv'
titanic_df = pd.read_csv("C:\\Users\\SAAD\\Desktop\\R n BI\\PYTHON\\titanic.csv")
```

```python
# Display the first few rows of the dataset
print(titanic_df.head())
```

```python
# Get information about the dataset
print(titanic_df.info())
```

```python
# Check for missing values
print(titanic_df.isnull().sum())
```

```python
# Assuming you want to fill missing age values with the median
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)
```

```python
#visualize data: - survival rate
sns.countplot(x='Survived', data=titanic_df)
plt.title('Survival Count')
plt.show()

```

```python
#demographics-Gender distribution
sns.countplot(x='Sex', data=titanic_df)
plt.title('Passenger Gender Distribution')
plt.show()

```

```python
#demographics - Passenger Class distribution
sns.countplot(x='Pclass', data=titanic_df)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
# age distribution:
sns.histplot(titanic_df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
#Survival by Class and Gender:
sns.countplot(x='Pclass', hue='Survived', data=titanic_df)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
#Survival by Class and Gender:
sns.countplot(x='Sex', hue='Survived', data=titanic_df)
plt.title('Survival by Gender')
plt.show()
```

```python
#Statistical Analysis:
#Example: Calculate Survival Rate by Class:
class_survival_rate = titanic_df.groupby('Pclass')['Survived'].mean()
print(class_survival_rate)
```

```python
# Conclusion
print("### Conclusion ###")
print("1. Survival Rate by Passenger Class:")
print(class_survival_rate)
```

```python
print("\n2. Gender Distribution:")
gender_distribution = titanic_df['Sex'].value_counts(normalize=True)
print(gender_distribution)
```

```python
# Explore Other Patterns or Characteristics
# For example, let's say you have a 'Fare' column indicating the ticket price:
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', data=titanic_df, hue='Survived', palette='coolwarm')
plt.title('Age vs. Fare with Survival Information')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
print("\n5. Fare and Age Relationship:")
print("   - There is a visible relationship between age, fare, and survival.")
print("   - Passengers who paid higher fares may have a higher likelihood of survival.")
```

```python
#more charts
#Correlation Heatmap:
# Generate a correlation matrix
# Drop non-numeric columns before calculating correlation
numeric_df = titanic_df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_df.corr()

```

```python
# Plot a heatmap to visualize correlations
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Create a pair plot to visualize relationships between numerical variables
sns.pairplot(titanic_df, hue='Survived', palette='viridis')
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()
```

```python
# Create box plots to visualize distributions and potential outliers
plt.figure(figsize=(12, 6))
sns.boxplot(x='Pclass', y='Age', data=titanic_df, hue='Survived', palette='Set2')
plt.title('Box Plot of Age by Passenger Class and Survival')
plt.show()
```

```python
# Create a violin plot to visualize the distribution of age by gender and survival
plt.figure(figsize=(10, 6))
sns.violinplot(x='Sex', y='Age', hue='Survived', data=titanic_df, split=True, palette='pastel')
plt.title('Violin Plot of Age by Gender and Survival')
plt.show()
```

```python
# Countplot for Siblings/Spouses Aboard
plt.figure(figsize=(10, 6))
sns.countplot(x='Siblings/Spouses Aboard', hue='Survived', data=titanic_df, palette='husl')
plt.title('Survival Count by Siblings/Spouses Aboard')
plt.show()



```

```python

# Countplot for Parents/Children Aboard
plt.figure(figsize=(10, 6))
sns.countplot(x='Parents/Children Aboard', hue='Survived', data=titanic_df, palette='muted')
plt.title('Survival Count by Parents/Children Aboard')
plt.show()
```
